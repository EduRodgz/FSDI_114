#Doubly Linked Lists

```python
class DoublyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None
            self.prev = None

    def __init__(self):
        self.head = None
        self.tail = None

    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % repr(current.data)
            current = current.next
            while current:
                out += ", %s" % repr(current.data)
            out += "]"
            return out
    
    def append(self, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            #current = sef.head
            #while current.next:
            #   current = current.next
            #current.next = new_code
            #new_node.prev = current
            #self.tail = new_node
            self.tail = new_node
            new_node.prev = self.tail

    def remove(self, value):
        #1.If head and tail node have 0 references the list is empty therefore empty state should return a IndexError("No values to remove")
        #2.If only one element is in list and it contains the target value it should be set to None ,if not the target value it should raise ValueError("Value does not exist")
        #3.If more than one element in list is the state, the list should be scanned until target value is found using a "current" to use references to go through the array to ensure every node is scanned for the target value, once target value is located the node has its references cut from the previous and next node in order to have the previous and next node connect skipping the target data node,this makes it so the target data node has 0 references and is  collected by garbage.
        #4. References should be updated when removing target data nodes, if target data node is removed the next node should change its reference to point to head, if target data node is the tail node reference and is removed the previous node should change its reference to tail.    
        #raise ValueError("Value does not exist")
        found = False
        if self.head:
            current = self.head
            while current and not found:
                if current.data == value:
                    found = True
                else:
                    current = current.next
            if found:
                if current == self.head:
                    self.head = self.head.next
                else:
                    prev_node = current.prev
                    next_node = current.next
                    prev_node.next = next_node
                    if next_node:
                        next_node.prev = prev_node
                    if current == self.tail:
                        self.tail = current.prev
        raise ValueError("Value does not exist")

    
    def insert(self, index, value):
        pass

    def __len__(self):
        current = self.head
        count = 0
        while current:
            current = current.next
            count += 1
        return count
```
